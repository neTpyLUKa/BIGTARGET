```python
def calc_corr(a, b):
    avg_a = np.mean(a)
    avg_b = np.mean(b)
    
    a_sub_avg = a - avg_a
    b_sub_avg = b - avg_b
    
    numerator = np.sum(a_sub_avg * b_sub_avg)
    
    denominator = (np.sum(a_sub_avg ** 2) * np.sum(b_sub_avg ** 2)) ** 0.5
    
    return numerator / denominator
```

```python
def features_order(df, n_sections):
    column_df = pd.DataFrame()
    
    column_df["column_name"] = list(df)
    
    correlations = []
    
    target = df["response_att"]
    
    for i, row in column_df.iterrows():
        column_name = row["column_name"]
        
        corr = calc_corr(target, df[column_name])
        
        correlations.append(abs(corr))
        
    column_df["corr"] = correlations
    
    return column_df
```

```python
column_df = features_order(pdf, 150)
```

```python
column_df.sort_values(by="corr", inplace=True, ascending=False)
```
